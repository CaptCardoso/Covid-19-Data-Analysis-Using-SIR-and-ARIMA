```python
import numpy as np
import pandas as pd
```

```python
df_oh = pd.read_csv('../data/thomas/df_oh_final2.csv')
df_oh.drop(columns=['administered_dose1_recip', 'series_complete_yes', 'new_dose_1', 'new_series_completed'], inplace=True)
```

```python
states = [ 'AK', 'AL', 'AR', 'AZ', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA',
           'HI', 'IA', 'ID', 'IL', 'IN', 'KS', 'KY', 'LA', 'MA', 'MD', 'ME',
           'MI', 'MN', 'MO', 'MS', 'MT', 'NC', 'ND', 'NE', 'NH', 'NJ', 'NM',
           'NV', 'NY', 'OK', 'OR', 'PA', 'RI', 'SC', 'SD', 'TN', 'TX',
           'UT', 'VA', 'VT', 'WA', 'WI', 'WV', 'WY']
```

```python
df_list = [df_oh]
for i in range(49):
    df = df_oh.copy()
    df['state'] = states[i]
    df['tot_cases'] = df_oh['tot_cases'] + 1000 * (i+1)
    df['new_cases'] = df_oh['new_cases'] + 100 * (i+1)
    df['I_actual'] = df_oh['I_actual'] + 500 * (i+1)
    df['I_arima_preds'] = df_oh['I_arima_preds'] + 250* (i+1)
    df['I_SIR_preds'] = df_oh['I_SIR_preds'] + 200 * (i+1)
    df_list.append(df)
```

```python
pd.concat(df_list).to_csv('../data/thomas/test.csv', index=False)
```

```python
df_all = pd.read_csv('../data/shaniaaaaaa/bangaroonie.csv', index_col='key_0').drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
df_all
```

```python
state_abbrev = [ 'AK_', 'AL_', 'AR_', 'AZ_', 'CA_', 'CO_', 'CT_', 'DC_', 'DE_', 'FL_',
                'GA_', 'IA_', 'ID_', 'IL_', 'IN_', 'KS_', 'KY_', 'LA_', 'MA_', 'MD_', 'ME_',
                'MI_', 'MN_', 'MO_', 'MS_', 'MT_', 'NC_', 'ND_', 'NE_', 'NH_', 'NJ_', 'NM_',
                'NV_', 'NY_', 'OH_', 'OK_', 'OR_', 'PA_', 'RI_', 'SC_', 'SD_', 'TN_', 'TX_', 'UT_',
                'VA_', 'VT_', 'WA_', 'WI_', 'WV_', 'WY_']

df_list = []

for abbrev in state_abbrev:
    state_cols = []
    for col in list(df_all.columns):
        if abbrev in col:
            state_cols.append(col)
#     print(state_cols)
    df = df_all[state_cols].drop(columns=[abbrev+'I_SRI',abbrev+'S_SRI']).copy()
#     print(df)
    df.columns = ['I', 'Vax', 'I_Arima', 'I_SIR', 'S_SIR', 'Herd_Immunity']
    df['state'] = abbrev[:2]
    df_list.append(df)
df_complete = pd.concat(df_list)
```
