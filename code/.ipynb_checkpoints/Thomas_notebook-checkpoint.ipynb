### ARIMA Model
I am starting with the total cases column, but will switch to new cases if the data is not stationary. I will be checking this with the Augmented Dickey-Fuller test, looking for the p-value to be less than .01. This will also provide the d value for the ARIMA model. 

```python
# Function pulled from the lesson-VAR-models starter-code notebook. Code written by Joseph Nelson and 
# improved by Hovanes Gasparian

def interpret_dftest(dftest):
    dfoutput = pd.Series(dftest[0:3], index=['Test Statistic','p-value', 'Lag Used'])
    return dfoutput
```

```python
interpret_dftest(adfuller(df_oh['tot_cases']))
# p-value not below .01, thus switching to 'new_cases' and applying .diff() from now on. 
```

```python
interpret_dftest(adfuller(df_oh['new_cases']))
```

```python
interpret_dftest(adfuller(df_oh['new_cases'].diff().dropna()))
```

Only needed one .diff() on the 'new_cases' column to reach stationarity. Thus the d value is 1. The first function below was taken from the lesson-ARIMA starter-code notebook that will find the best p and q values for the ARIMA model. The second function, also from the lesson-ARIMA starter-code notebook, will generate a prediction for each day in the test data and plot them against the actual values and the training data.Please ignore all the warnings.

```python
train, test = train_test_split(df_oh['new_cases'], test_size=.1, shuffle=False)
```

```python
def best_p_q(d_value=1, range_val=5, train_data=train):
    # https://queirozf.com/entries/suppressing-ignoring-warnings-in-python-reference-and-examples for 
    # help with hiding the convergence and value warnings that happen and can be ignored
    with warnings.catch_warnings():
        warnings.simplefilter('ignore')
        best_aic = 99999999
        best_p = 0
        best_q = 0

        for p in range(range_val):
            for q in range(range_val):
                try:
                    arima = ARIMA(endog=train_data, order=(p, d_value, q))
                    model = arima.fit()

                    if model.aic < best_aic:

                        best_aic = model.aic
                        best_p = p
                        best_q = q

                except:
                    pass
        print('MODEL FINISHED!')
        print(f'Our model that minimizes AIC on the training data is the ARIMA({best_p},{d_value},{best_q}).')
        print(f'This model has an AIC of {best_aic}.')
        return (best_p, d_value, best_q)
    
```

```python
def plot_ARIMA(train_data=train, test_data=test, p=0, d=0, q=0):
    with warnings.catch_warnings():
        warnings.simplefilter('ignore')
        arima = ARIMA(endog=train_data, order=(p,d,q))
        model = arima.fit()

        preds = model.predict(start=test_data.index[0], end=test_data.index[-1])
        preds

        plt.figure(figsize=(10,6))
        plt.plot(train_data.index, train_data, color = 'blue', label='Training Data')
        plt.plot(test_data.index, test_data, color = 'orange', label='Actual Cases')
        plt.plot(test_data.index, preds, color = 'green', label='Predicted Cases')
        plt.title(label = 'ARIMA(8,1,6) Predictions', fontsize=16)
        plt.legend()
        plt.show();
```

```python
p_val, d_val, q_val= best_p_q(d_value=1, range_val=10, train_data=train)
```

```python
plot_ARIMA(p=p_val, d=d_val, q=q_val, train_data=train, test_data=test)
```

As shown by the above model, ARIMA is pretty decent for a couple days, but quickly bounces back to the mean, which is not close to the true data. I am going to rerun the same models below showing if it was run on total cases and at 2 .diffs().

```python
train_tot, test_tot = train_test_split(df_oh['tot_cases'], test_size=.1, shuffle=False)
```

```python
p_val, d_val, q_val= best_p_q(d_value=1, range_val=10, train_data=train_tot)
```

```python
plot_ARIMA(p=p_val, d=d_val, q=q_val, train_data=train_tot, test_data=test_tot)
```

This shows that ARIMA is pretty decent for a couple days but then it returns to the mean and goes up to a constant rate while the real data goes up at a nearly quadratic rate

### SIR Model
The first function, developed by David, sets up the SIR model function. I will then be putting in Ohio's data to run the model. The population of Ohio came from the [U.S. Census](https://www.census.gov/library/stories/state-by-state/ohio-population-change-between-census-decade.html) and is the variable 'pop' below. This is also being run on the total cases column.

```python
def SIR_sim(S0=.9,
            I0=.1,
            d = 100, 
            p = 1, 
            a = 10**(-3),
            b = 10**(-1), 
            c = 0,
            alpha = 0, 
            dt=1):
    """Takes initial populations and predicts populations at all later times.
    The function is defined with default values to enable group member to play with it."""
    # Initialize S,I,R
    S = pd.Series(data = np.zeros(d,))
    I = pd.Series(data = np.zeros(d,))
    R = pd.Series(data = np.zeros(d,))
    S[0] = S0
    I[0] = I0
    R[0] = p-S0-I0

    # calculate later values from initial values
    for t in range(0,d):
        dS = -a*S[t]*I[t]           * dt #ds/dt = -aSI 
        dI = (a*S[t]*I[t] - b*I[t]) * dt # dI/dt = aSI-bI
        dR = b*I[t]                 * dt 
        S[t+1] = S[t]+dS
        I[t+1] = I[t]+dI
        R[t+1] = R[t]+dR

    # collect into a data frame
    columns_SIR = ['day','S','I','R']
    eg = pd.DataFrame(columns = columns_SIR)
    eg['day'] = range(d+1)
    eg['S'] = S
    eg['I'] = I
    eg['R'] = R
    return eg
```

```python
pop = 11_799_448
days = len(df_oh) # Number of days is length of list
b = 1/14 # in case we decide this is wrong. 

I0 = df_oh['tot_cases'][0]
eg = SIR_sim(d = days,
             p = pop,
             I0 = I0,
             S0 = 0.99*(pop-I0) , #0.9*(p-I0) assumes 10% of the uninfected pop is recovered initially
             a = (2.8*b)/pop,  
             b = b 
            )


plt.plot(eg['day'], eg['S'],label='S')
plt.plot(eg['day'], eg['I'], label='I')
plt.plot(eg['day'], eg['R'], label ='R')
plt.xlabel('Day')
plt.ylabel('Counts')
plt.legend();
```

### Rerunning ARIMA for Day to Day Predictions 
In this section, we will be running ARIMA and getting the next day's prediction, retraining it, getting the next day's prediction, refitting, so on and so forth until we have predictions for every day in our collected dataset. This function was created by David.

```python
df_oh['I_actual'] = df_oh['new_cases'].rolling(14).sum()
```

```python
with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    df_oh['I_actual'].iloc[0:13] = df_oh['tot_cases'].iloc[:13]
```

```python
df_oh['I_actual'].iloc[0:13] 
```

```python
df_oh.head(15)
```

```python
with warnings.catch_warnings():
    warnings.simplefilter('ignore')
    y = df_oh['I_actual'].copy()
    I = pd.Series(np.zeros((len(df_oh),)))

    p = 1
    I[0:p] = df_oh['I_actual'].copy()[0:p]

    for day in range(p+2,len(df_oh)-1 ):
        arima = ARIMA( endog = y[y.index[0:day]], order =(p,0,0) )      
        model = arima.fit()
        I[day+1] = model.predict( start = y.index[day+1], end = y.index[day+1]   )
    df_oh['I_arima_preds'] = list(I) # This code adds the ARIMA predictions to the dataframe
```

```python
plt.figure(figsize=(10,10))
plt.plot(I, label='Predicted I')
plt.plot(y.values, label='Actual I')
plt.title('Arima(1,0,0) 1 Day Out Predictions of Infections')
plt.xlabel('Day')
plt.ylabel('Infected')
plt.text(x=100,y=200_000, s= f"MAE={round(mean_absolute_error(I,df_oh['I_actual']  ),3)}")
plt.legend();
```

# Rerunning the SIR to get Daily Predictions
Rerunning David's SIR model to get daily predictions for new cases.

```python
def next_day_SI_withC(S,I):
    '''Predicts the number of infected tomorrow and number of susceptable tomorrow
    from number of infrected today 
    and number susceptable today'''
    #parameters 
    pop = 11_799_448
    b = 1/14 # Recovery rate for duration 14 days
    a = 2.8*b/pop # transmission rate for reproduction number 2.8 and recovery rate 1/14
    c = 1/100 # rate of immunity loss
    
    S_next = S - a*S*I +c*(p-S-I)
    I_next = I + a*S*I -b*I
    return S_next,I_next

# initialize 
S = pd.Series( np.zeros((len(df_oh),)) )
I = pd.Series( np.zeros((len(df_oh),)) )
best_r = 0.8887 # Initial susceptable ratio
S[0] = best_r*pop # initial susceptable population
I[0] = df_oh['I_actual'].copy()[0]

for day in range(0,len(df_oh) -1):
    S[day+1], I[day+1] = next_day_SI_withC(S[day],df_oh['I_actual'][day])
    
df_oh['I_SIR_preds'] = list(I) # This code adds the SIR predictions to the dataframe
```

```python
# plt.plot(S,label='S')
plt.plot(I, label='Predicted I')
plt.plot(range(len(df_oh)) , df_oh['I_actual'] , label ='Actual I')
plt.title('SIR 1 Day Out Predictions of Infections')
plt.xlabel('Day')
plt.ylabel('Counts')
plt.text(x=100,y=200_000, s= f"MAE={round(mean_absolute_error(I,df_oh['I_actual']  ),3)}")
plt.legend();
```
